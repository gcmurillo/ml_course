# Cargando datos con open

Cuando el dataset es de un tamano significativo, no es recomendable cargarlo con pandas, dado que estos cargan todos los datos en memoria, con open podemos tener los datos por trozos o linea por linea.

```python
data2 = pd.read_csv(main_path + '/' + 'customer-churn-model/Customer Churn Model.txt')
```

```python
data2.head()
```

```python
data3 = open(main_path + '/' + 'customer-churn-model/Customer Churn Model.txt', 'r')
```

```python
cols = data3.readline().strip().split(',')
n_cols = len(cols)
```

```python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
main_dict
```

```python
for line in data3:
    values = line.strip().split(',')
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1
    
print("El dataset tiene %d filas, %d columnas" % (counter, n_cols))
```

```python
df3 = pd.DataFrame(main_dict)
df3.head()
```

### Creando diccionario de datos desde url y transformandolo a DataFrame

```python
# Getting data for url
import urllib3 
http = urllib3.PoolManager()
r = http.request('GET', medals_url)
r.status
response = r.data

# Decoding
str_data = response.decode('utf-8')

# Dividir las lineas
lines = str_data.split('\n')
cols_names = lines[0].split(',')
n_cols = len(cols)
files = 0

# Creando diccionario de datos
main_dict = {}
for col in cols_names:
    main_dict[col] = []

for line in range(1, len(lines)):
    data_line = lines[line].split(',')
    for i in range(len(data_line)):
        main_dict[cols_names[i]].append(data_line[i])
    files += 1

# Presentando lineas y creando DataFrame
print("El dataset tiene %d filas y %d columnas" % (files, n_cols))
df4 = pd.DataFrame(main_dict)
print(df4.head())

main_path = '../datasets'
dataset_path = 'athletes/downloaded_medals.'
fullpath = os.path.join(main_path, dataset_path)

df4.to_csv(fullpath+'csv')
df4.to_json(fullpath+'json')
```
